```python

#quicksort is unstable,inplace algo,comparison sort,recursive algo,also called partiton exchange sort,
# while comparing pivot should be shifted with right when it is chosen at first 
#3 times faster than merge ,heap sort
def pivot_place(list1, first, last):
    pivot = list1[last]
    right = last - 1
    left = first
    
    while True:
        while left <= right and list1[left] <= pivot:
            left += 1
        print(list1)
        while left <= right and list1[right] >= pivot:
            right -= 1
        print(list1)
        if left > right:
            break
        else:
            list1[left], list1[right] = list1[right], list1[left]
        
    list1[last], list1[left] = list1[left], list1[last]
    return left

#manupulating in the list itself so no need to return
def quicksort(list1,first,last):
    if first<last:
        p=pivot_place(list1,first,last)
        quicksort(list1,first,p-1)
        quicksort(list1,p+1,last)

    
list1=[56,26,93,17,31,44]
n=len(list1)
quicksort(list1,0,n-1)
print(list1)
        
```

```python
# Python program for implementation of MergeSort

# Merges two subarrays of arr[].
# First subarray is arr[l..m]
# Second subarray is arr[m+1..r]


def merge(arr, l, m, r):
	n1 = m - l + 1
	n2 = r - m

	# create temp arrays
	L = [0] * (n1)
	R = [0] * (n2)

	# Copy data to temp arrays L[] and R[]
	for i in range(0, n1):
		L[i] = arr[l + i]

	for j in range(0, n2):
		R[j] = arr[m + 1 + j]

	# Merge the temp arrays back into arr[l..r]
	i = 0	 # Initial index of first subarray
	j = 0	 # Initial index of second subarray
	k = l	 # Initial index of merged subarray

	while i < n1 and j < n2:
		if L[i] <= R[j]:
			arr[k] = L[i]
			i += 1
		else:
			arr[k] = R[j]
			j += 1
		k += 1

	# Copy the remaining elements of L[], if there
	# are any
	while i < n1:
		arr[k] = L[i]
		i += 1
		k += 1

	# Copy the remaining elements of R[], if there
	# are any
	while j < n2:
		arr[k] = R[j]
		j += 1
		k += 1

# l is for left index and r is right index of the
# sub-array of arr to be sorted


def mergeSort(arr, l, r):
	if l < r:

		# Same as (l+r)//2, but avoids overflow for
		# large l and h
		m = l+(r-l)//2

		# Sort first and second halves
		mergeSort(arr, l, m)
		mergeSort(arr, m+1, r)
		merge(arr, l, m, r)


# Driver code to test above
arr = [12, 11, 13, 5, 6, 7]
n = len(arr)
print("Given array is")
for i in range(n):
	print("%d" % arr[i],end=" ")

mergeSort(arr, 0, n-1)
print("\n\nSorted array is")
for i in range(n):
	print("%d" % arr[i],end=" ")

# This code is contributed by Mohit Kumra
```

```python
# Algorithm for insertion sorting:
# 1. consider the first element to be sorted and the rest to be unsorted.
# 2. take the first element in the unsorted part(u1) and compare it with sorted part elements(s1).
# 3. if u1<s1 then insert u1 in the correct index, else leave it as it is.
# 4. take next element in the unsorted part and
##card placement

def insertionsort(l):
    for i in range(1,len(l)):
        current_element=l[i]
        pos=i
        while current_element<l[pos-1] and pos>0:
            l[pos-1],l[pos]=l[pos],l[pos-1]#we are shifting the indexed element only means pos=current checking elent position
            pos=pos-1
        l[pos]=current_element
    
    
l=[2,4,3,5,1]
# l=[5,10,9,7,78,90]

insertionsort(l)
print(l)
```

```python
##binary search

def bs(s,n):
    low=0
    high=len(s)-1
    found=False
    while low<=high and not found:
        mid=(low+high)//2
        if s[mid]==n:
            found=True
    
        elif n>s[mid]:
            low=mid+1
        else:
            high=mid-1
    if found==True:
        print("the element is found at ",mid)
        
    else:
        print("the element you are searching didnt found")
    

l=[1,2,3,4,5,6,7,8,9,0,12,13,14,15,13,13,15]
s=sorted(l)
n=14
print(s)
bs(s,n)
```

```python
#BST
class BST():
    def __init__(self,key):#for creation of new node
        self.key=key
        self.lchild=None
        self.rchild=None
    def insert(self,data):
        if self.key is None:
            self.key=data
            return
        if self.key==data:#duplicate ignore
            return
        if data<self.key:
            if self.lchild:
                self.lchild.insert(data)
            else:
                self.lchild=BST(data)
        else:
            if self.rchild:
                self.rchild.insert(data)
            else:
                self.rchild=BST(data)    
    def search(self,data):
        if self.key==data:
            print("the Node is Found:")
            return
        if data<self.key:
            if self.lchild:
                self.lchild.search(data)
            else:
                print("the node is not present")
        else:
            if self.rchild:
                self.rchild.search(data)
            else:
                print("the node is not present")
                
    def preorder(self):
        print(self.key,end=" ")
        if self.lchild:
            self.lchild.preorder()
        if self.rchild:
            self.rchild.preorder()
    
    def inorder(self):
        if self.lchild:
            self.lchild.inorder()
        print(self.key,end=" ")
        if self.rchild:
            self.rchild.inorder()
    
    def postorder(self):
        if self.lchild:
            self.lchild.postorder()
        if self.rchild:
            self.rchild.postorder()
        print(self.key,end=" ")
    
#while deleting the node:replace the parent node with largest key in the left subtree or with the smallest key in the right subtree
    def delete(self,data):
        if self.key is None:
            print("the tree is empty")
        if data<self.key:
            if self.lchild:
                self.lchild=self.lchild.delete(data)
            else:
                print("the given node not present")
        elif data>self.key:
            if self.rchild:
                self.rchild=self.rchild.delete(data)
            else:
                print("the tree is empty")
        else:
            if self.rchild is None:
                temp=self.lchild
                self=None
                return temp
            if self.lchild is None:
                temp=self.rchild
                self=None
                return temp
            node=self.rchild
            while self.lchild:
                node=self.lchild
            self.key=node.key
            self.rchild=self.rchild.delete(node.key)
        return self
# # Definition for a binary tree node.
# # class TreeNode:
# #     def __init__(self, val=0, left=None, right=None):
# #         self.val = val
# #         self.left = left
# #         self.right = right
# class Solution:
#     def deleteNode(self, root: Optional[TreeNode], key: int) -> Optional[TreeNode]:
#         if not root:
#             return root
        
#         if key > root.val:
#             root.right = self.deleteNode(root.right, key)
#         elif key < root.val:
#             root.left = self.deleteNode(root.left, key)
#         else:
#             if not root.left:
#                 return root.right
#             elif not root.right:
#                 return root.left
            
#             # Find the min from right subtree
#             cur = root.right
#             while cur.left:
#                 cur = cur.left 
#             root.val = cur.val
#             root.right = self.deleteNode(root.right, root.val)
#         return root
               
                
root=BST(10)
# root.insert(20)
l=[6,3,1,6,98,3,7]
for i in l:
    root.insert(i)
root.search(1)
print("preorder")
root.preorder()
print()
print("inorder")
root.inorder()
print()
print("postorder")
root.postorder()
root.delete(98)

print("\nafter deleting:")
root.postorder()
```
