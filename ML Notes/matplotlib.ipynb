```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
print(x)
```

```python
y=2*x
print(y)
```

```python

plt.subplot(1,2,1)
plt.plot(x,y,color="g",linestyle=":",linewidth=4)
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(y,x,color="b",linestyle=":",linewidth=4)
plt.grid(True)


plt.title("vasu")
plt.xlabel("hmm")
plt.ylabel("hello")

plt.show()
```

```python
plt.plot(x,y,color="g",linestyle="-",linewidth=2)
plt.plot(y,x,color="b",linestyle="-",linewidth=3)

```

```python
student={"vasu":70,"Hemanth":80,"monesh":90,"pspk":100}
```

```python
keys=list(student.keys())
values=list((student.values()))
plt.bar(keys,values)
plt.grid(True)
plt.show
```

```python
keys=list(student.keys())
values=list((student.values()))
plt.barh(keys,values,color="g")
plt.grid(True)
plt.show
```
